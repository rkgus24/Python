기상 관측 이래, 서울의 최고 기온이 가장 높았던 날은 언제였고, 몇 도였을까?

#### 6. 문제 해결 방법 구상하기
- 1) 데이터를 읽어온다.
- 2) 순차적으로 최고 기온을 확인한다.
- 3) 최고 기온이 가장 높았던 날짜의 데이터를 저장한다.
- 4) 최종 저장된 데이터를 출력한다.

```python
# 2) 순차적으로 최고 기온을 확인한다
filename = '/content/drive/MyDrive/DATA/busan.csv'
import csv
f = open(filename, 'r' , encoding='cp949')
data = csv.reader(f)
header = next(data)
for row in data:
  print(row[2])
f.close()
```

```python
# 에러해결 - 빈 데이터로 인한 예간
filename = '/content/drive/MyDrive/DATA/busan.csv'
import csv
f = open(filename, 'r' , encoding='cp949')
data = csv.reader(f)
header = next(data)
for row in data:
  if row[2] =='':
    row[2] = -999
  print(row[2])
f.close()
```

```python
# 부산의 기온이 가장 높았던 날의 날짜와 기온 구하기
filename = '/content/drive/MyDrive/DATA/busan.csv'
import csv
f = open(filename, 'r', encoding='UTF-8')
data = csv.reader(f)
header = next(data)
max_temp = -999
max_data=''
for row in data:
  if row[2] == '':
    row[2] = -999
  row[2] = float(row[2])
  if max_temp < row[2]:
    max_data = row[0]
    max_temp = row[2]
f.close()
print(max_data, max_temp)
```

```python
# 부산의 기온이 가장 낮았던 날의 날짜와 기온 구하기
filename = '/content/drive/MyDrive/DATA/busan.csv'
import csv

f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
header = next(data)

min_temp = 999
min_data = ''

for row in data:
    if row[3] == '':
        row[3] = -999
    row[3] = float(row[3])
    
    if min_temp > row[3]:
        min_data = row[0]
        min_temp = row[3]

f.close()

print(min_data, min_temp)
```
