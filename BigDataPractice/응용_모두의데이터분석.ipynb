```python
# 부산 해운대구 우제1동의 0~100세 이상까지의 인구수를 순서대로 저장
# 문제
filename = '/content/drive/MyDrive/DATA/age.csv'
import csv
f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
for row in data:
  if '우제1동' in row[0]: # <-- 수정 : 0~100세까지 출력
    print(row)
f.close()
```

```python
# 부산 해운대구 우제1동의 0~100세 이상까지의 인구수를 순서대로 저장
# 가효링이 풀어본 문제
filename = '/content/drive/MyDrive/DATA/age.csv'
import csv
f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
result = [] # 결과를 저장할 리스트
for row in data:
    if '우제1동' in row[0]:
        for i in range(3, 104): # 0~100세 이상까지
            result.append(int(row[i].replace(',', ''))) # 숫자로 변환하여 리스트에 저장
print(result)
f.close()
```

```python
# 부산 해운대구 우제1동의 0~100세 이상까지의 인구수를 순서대로 저장
# 쌤이 푼 문제
filename = '/content/drive/MyDrive/DATA/age.csv'
import csv
f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
for row in data:
  if '우제1동' in row[0]: # <-- 수정 : 0~100세까지 출력
  # print(row[3:])
    for i in row[3:]:
      print(i, end=' ')
f.close()
```

```python
# 읽어온 데이터 저장
# 저장할 때는 리스트 변수를 사용하는 것이 좋음
filename = '/content/drive/MyDrive/DATA/age.csv'
import csv
f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
result = []
for row in data:
  if '우제1동' in row[0]:
  # print(row[3:])
    for i in row[3:]:
      result.append(i)  # <--- 수정 리스트에 저장
f.close()
print(result)
```

```python
# 문자열 출력값을 그래프를 위해서 수치형으로 바꾼다
# 문제 : 그래프로 출력
# 가효링이 푼 문제
import csv
import matplotlib.pyplot as plt
filename = '/content/drive/MyDrive/DATA/age.csv'

f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)

result = []
for row in data:
  if '우제1동' in row[0]:
    plt.show()
  # print(row[3:])
    for i in row[3:]:
      result.append(i) 
f.close()

x = list(range(101))
y = [int(data) for data in result]
plt.plot(x, y)
plt.xlabel('나이')
plt.ylabel('분포')
plt.title('헤헤티비')
plt.show()
```

```python
# 문자열 출력값을 그래프를 위해서 수치형으로 바꾼다
# 부산광역시 금정구 장전제1동(2641060000)
# 문제 : 읍면동 검색해서 그래프로 출력
import csv
import matplotlib.pyplot as plt
filename = '/content/drive/MyDrive/DATA/age.csv'

f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)

result = []
for row in data:
  if '우제1동' in row[0]:
    plt.show()
  # print(row[3:])
    for i in row[3:]:
      result.append(i) 
f.close()

x = list(range(101))
y = [int(data) for data in result]
plt.plot(x, y)
plt.xlabel('나이')
plt.ylabel('분포')
plt.title('헤헤마라탕먹고싶다집가서놀고싶다용가리치킨존맛탱구리티비')
plt.show()
```

```python
filename='/content/drive/MyDrive/DATA/age.csv'
import csv
f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
age_list = []
local = input('지역을 입력해주세요 : ')
for row in data:
  if local in row[0]:
    for i in row[3:]:
      age_list.append(int(i))
f.close()
print(age_list)
plt.plot(age_list)
plt.show()
```
